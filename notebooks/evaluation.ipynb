# Architecture Evaluation: Multi-Agent LLMs vs. Classical TF-IDF

In this notebook, we evaluate our LangGraph GenAI pipeline against a classical NLP baseline (TF-IDF Cosine Similarity) to demonstrate the qualitative improvement in resume screening.

```python
import sys
sys.path.append('..')
from baselines.tfidf_ranker import TfidfRanker
from utils import extract_text_from_pdf
import os

jd = "Machine Learning Engineer with Deep Learning and PyTorch"
files = [os.path.join('../batch_pdfs', f) for f in os.listdir('../batch_pdfs') if f.endswith('.pdf')][:3]
texts = [extract_text_from_pdf(f) for f in files]
```

## 1. Classical Approach: TF-IDF + Cosine Similarity
TF-IDF captures strict keyword matching. It lacks semantic understanding (e.g., treating 'Computer Vision' and 'Image Processing' as fully distinct).

```python
ranker = TfidfRanker()
scores = ranker.rank_resumes(jd, texts)
for f, s in zip(files, scores):
    print(f"{os.path.basename(f)} Score: {s:.3f}")
```

## 2. Multi-Agent Approach (LangGraph)
By passing the text through our routing classifier, to an extractor, and ultimately to an `AnalystAgent`, we achieve contextual reasoning. The Agent yields a qualitative SWOT alongside a calculated score, drastically outperforming pure keyword cosine metrics on qualitative resumes.
